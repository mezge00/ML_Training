```python
import pandas as pd
import sklearn 
from sklearn import svm, preprocessing 
df=pd.read_csv("datasets/diamonds.csv", index_col=0)
df.head()
```

```python
df["cut"].unique()
```

```python
#df["cut"].astype("category").cat.codes
cut_class_dict={"Fair":1, "Good":2, "Very Good":3, "Premium":4, "Ideal":5}
color_class_dict={"J":1, "I":2, "H":2, "G":3, "F":4, "E":5,"D":6}
clarity_class_dict={"I1":1, "SI2":2, "SI1":3, "VS2":4, "VS1":5, "VVS2":6, "VVS1":7, "IF":8}
```

```python
df["cut"]=df["cut"].map(cut_class_dict)
df["color"]=df["color"].map(color_class_dict)
df["clarity"]=df["clarity"].map(clarity_class_dict)
```

```python
df.head()
```

```python
df=sklearn.utils.shuffle(df)
X=df.drop("price",axis=1).values
Y=df["price"].values
```

```python
X=preprocessing.scale(X)
test_size=200
x_train=X[:-test_size]
y_train=Y[:-test_size]
x_test=X[-test_size:]
y_test=Y[-test_size:]

```

```python
clf=svm.SVR(kernel="sigmoid")
clf.fit(x_train,y_train)
```

```python
clf.score(x_test,y_test)
```

```python
for X,y in zip(x_test,y_test):
    print(f"Model:{clf.predict([X])[0]},Actual:{y}")
```
